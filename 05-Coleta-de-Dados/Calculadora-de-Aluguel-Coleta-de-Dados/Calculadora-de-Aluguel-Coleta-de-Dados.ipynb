```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import time

# Functions
def convert_df(area_lista, endereco_lista, valor_lista, quarto_lista, banheiro_lista, vaga_lista):
    data = {'area': area_lista,
            'endereco':endereco_lista,
            'aluguel': valor_lista,
            'quartos': quarto_lista,
            'banheiros': banheiro_lista,
            'vagas': vaga_lista}
    df = pd.DataFrame(data)

    return df

def scrap(url):
    c=requests.get(url).content
    soup=BeautifulSoup(c)

    area_tags = soup.find_all('span', {'class':'property-card__detail-value js-property-card-value property-card__detail-area js-property-card-detail-area'})   
    area_lista=[ar.text for ar in area_tags]
        
    address_tags = soup.find_all('span', {'class':'property-card__address'})
    endereco_lista=[en.text for en in address_tags]

    price_tags = soup.find_all('div', {'class':'property-card__price js-property-card-prices js-property-card__price-small'})
    valor_lista = [vl.text for vl in price_tags]    
   
    room_tags = soup.find_all('li', {'class':'property-card__detail-item property-card__detail-room js-property-detail-rooms'})   
    quarto_lista=[qt.text for qt in room_tags]
    
    bath_tags = soup.find_all('li', {'class':'property-card__detail-item property-card__detail-bathroom js-property-detail-bathroom'})  
    banheiro_lista=[ba.text for ba in bath_tags]

    garage_tags = soup.find_all('li', {'class':'property-card__detail-item property-card__detail-garage js-property-detail-garages'})  
    vaga_lista=[vg.text for vg in garage_tags]
    
    df=convert_df(area_lista, endereco_lista, valor_lista, quarto_lista, banheiro_lista, vaga_lista)

    return df
```

```python
qtpaginas = 3
URL = 'https://www.vivareal.com.br/aluguel/sp/sao-caetano-do-sul/'

paginas = ['']
for i in range(2,qtpaginas):
    a = '?pagina-'+str(i)
    paginas.append(a)

df_lista = []
for pg in paginas:
    #url=URL.format(pg)
    url = URL+pg
    print('Coletando URL:', url)
    try:
        df=scrap(url)
    except Exception as e:
        print('Erro ao coletar URL:', url)
        print(e)
    df_lista.append(df)
    time.sleep(1)
df_final = pd.concat(df_lista)
```
